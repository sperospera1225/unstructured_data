```python
training_text = [
    "I enjoy reading books in my free time.",
    "She is a talented musician and plays the piano beautifully.",
    "The sunsets at the beach are always breathtaking.",
    "Learning a new language can be challenging but rewarding.",
    "He completed his homework before going out to play.",
    "Cooking homemade meals is a great way to relax for me.",
    "The city skyline looked magical under the night sky.",
    "I prefer to drink tea instead of coffee in the morning.",
    "They built a new playground for the children in the neighborhood.",
    "I love listening to classical music while I work.",
    "Swimming in the ocean makes me feel incredibly refreshed.",
    "The smell of freshly baked bread is irresistible.",
    "My best friend and I have known each other since kindergarten.",
    "Traveling to different countries allows you to experience diverse cultures.",
    "Reading poetry can be a deeply moving and emotional experience.",
    "The little girl giggled as the puppy licked her face.",
    "He carefully planted flowers in the garden, creating a colorful display.",
    "The old bookstore in the corner of the street has a unique charm.",
    "Watching a comedy movie is a great way to unwind after a long day.",
    "She was captivated by the beauty of the blooming flowers in the park.",
    "I believe that kindness and empathy can change the world.",
    "Learning to play a musical instrument requires dedication and practice.",
    "The smell of rain on the earth is one of my favorite scents.",
    "He excels in mathematics and often helps his classmates with difficult problems.",
    "Exploring nature trails is a wonderful way to connect with the outdoors.",
    "She carefully decorated her room with fairy lights and colorful posters.",
    "I find inspiration for my artwork in everyday life and nature.",
    "The laughter of children playing in the park is contagious.",
    "Waking up early allows me to enjoy the peacefulness of the morning.",
    "Writing in a journal helps me organize my thoughts and feelings.",
    "The aroma of freshly brewed coffee fills the air in the café.",
    "He enthusiastically participates in community service projects.",
    "Dancing to music is a great way to express oneself creatively.",
    "A warm hug from a friend can instantly improve my mood.",
    "I am fascinated by the wonders of the universe and enjoy stargazing.",
    "She bakes delicious cookies and shares them with her neighbors.",
    "Reading historical novels transports me to different eras in time.",
    "Spending time with my grandparents is always a cherished experience.",
    "The sound of waves crashing on the shore is incredibly calming.",
    "The old lighthouse stood tall against the backdrop of the stormy sea.",
    "He volunteered at the animal shelter, caring for abandoned pets.",
    "Painting landscapes allows me to capture the beauty of nature on canvas.",
    "The scent of pine trees in the forest is invigorating and fresh.",
    "Attending live concerts gives me an adrenaline rush and fills me with energy.",
    "She planted a variety of herbs in her garden for culinary use.",
    "The city comes alive with colorful lights during the holiday season.",
    "Building a sandcastle on the beach is a fun and creative activity.",
    "My favorite childhood memory is baking cookies with my grandmother.",
    "The melody of the violin always brings tears to my eyes.",
    "Volunteering at the local soup kitchen taught me the value of compassion and giving."
]
```

```python
test_sentences = [
    "I enjoy reading",
    "She is a talented",
    "Cooking homemade meals is a great",
    "The little girl giggled",
    "Solving puzzles challenges my"
]
```

```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, GRU, Dense, SimpleRNN


# Tokenizer 객체 생성
tokenizer = Tokenizer()
tokenizer.fit_on_texts(training_text)

# 텍스트를 정수 시퀀스로 변환
sequences = tokenizer.texts_to_sequences(training_text)

# 입력 시퀀스와 레이블 생성
input_sequences = []
output_sequences = []

for sequence in sequences:
    for i in range(1, len(sequence)):
        input_sequences.append(sequence[:i])
        output_sequences.append(sequence[i])

# 입력 시퀀스를 패딩하여 동일한 길이로 만듦
max_sequence_length = max([len(seq) for seq in input_sequences])
input_sequences = pad_sequences(input_sequences, maxlen=max_sequence_length, padding='pre')

# 레이블을 one-hot 인코딩
num_words = len(tokenizer.word_index) + 1
output_sequences = np.eye(num_words)[output_sequences]
```

```python
# RNN 모델 학습
rnn_model = Sequential()
rnn_model.add(Embedding(input_dim=num_words, output_dim=50, input_length=max_sequence_length))
rnn_model.add(SimpleRNN(units=100, activation='tanh'))
rnn_model.add(Dense(units=num_words, activation='softmax'))

rnn_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
rnn_model.fit(input_sequences, output_sequences, epochs=50, batch_size=32, validation_split=0.2)
```

```python
# LSTM 모델 학습
lstm_model = Sequential()
lstm_model.add(Embedding(input_dim=num_words, output_dim=50, input_length=max_sequence_length))
lstm_model.add(LSTM(units=100))
lstm_model.add(Dense(units=num_words, activation='softmax'))

lstm_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
lstm_model.fit(input_sequences, output_sequences, epochs=50, batch_size=32, validation_split=0.2)
```

```python
# GRU 모델 학습
gru_model = Sequential()
gru_model.add(Embedding(input_dim=num_words, output_dim=50, input_length=max_sequence_length))
gru_model.add(GRU(units=100))
gru_model.add(Dense(units=num_words, activation='softmax'))

gru_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
gru_model.fit(input_sequences, output_sequences, epochs=50, batch_size=32, validation_split=0.2)
```

```python
# 테스트 문장을 정수 시퀀스로 변환
test_sequences = tokenizer.texts_to_sequences(test_sentences)
test_sequences = pad_sequences(test_sequences, maxlen=max_sequence_length, padding='pre')

# 각 모델에서 다음 토큰 예측
predicted_rnn = rnn_model.predict(test_sequences)
predicted_lstm = lstm_model.predict(test_sequences)
predicted_gru = gru_model.predict(test_sequences)

# 예측 결과를 텍스트로 변환
predicted_rnn_texts = [tokenizer.index_word.get(np.argmax(sequence), "") for sequence in predicted_rnn]
predicted_lstm_texts = [tokenizer.index_word.get(np.argmax(sequence), "") for sequence in predicted_lstm]
predicted_gru_texts = [tokenizer.index_word.get(np.argmax(sequence), "") for sequence in predicted_gru]

# 예측 결과 출력
for i in range(len(test_sentences)):
    print(f"Input: {test_sentences[i]}")
    print(f"Predicted Next Token (RNN): {predicted_rnn_texts[i]}")
    print(f"Predicted Next Token (LSTM): {predicted_lstm_texts[i]}")
    print(f"Predicted Next Token (GRU): {predicted_gru_texts[i]}")
    print("-----------------------")
```
